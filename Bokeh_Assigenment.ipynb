```python
from bokeh.plotting import figure, output_file, show

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Output to static HTML file (optional)
output_file("scatter_plot.html")

# Create a new plot with a title and axis labels
p = figure(title="Simple Scatter Plot", x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a circle glyph with size and color
p.circle(x, y, size=10, color="navy", alpha=0.5)

# Show the plot
show(p)
```

```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
"""In Bokeh, glyphs are the basic visual building blocks used to represent data in plots. They are geometric shapes like circles, squares, lines, and more, that can be added to a plot to visually represent data points, lines, bars, or other graphical elements.

To add glyphs to a Bokeh plot, you typically use methods provided by the figure object. These methods allow you to specify the type of glyph (circle, square, line, etc.), along with attributes such as size, color, and alpha (transparency), and then add them to the plot."""
```

```python
from bokeh.plotting import figure, output_file, show

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Output to static HTML file
output_file("glyphs_example.html")

# Create a new plot with a title and axis labels
p = figure(title="Glyphs Example", x_axis_label='X-axis', y_axis_label='Y-axis')

# Add circle glyph
p.circle(x, y, size=10, color="navy", alpha=0.5, legend_label="Circle")

# Add square glyph
p.square(x, y, size=10, color="firebrick", alpha=0.5, legend_label="Square")

# Add line glyph
p.line(x, y, line_width=2, color="green", alpha=0.5, legend_label="Line")

# Add text glyph
p.text(x, y, text=["Point 1", "Point 2", "Point 3", "Point 4", "Point 5"], text_baseline="middle", text_align="center", text_font_size="10pt")

# Add legend
p.legend.location = "top_left"

# Show the plot
show(p)
```

```python
#Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
from bokeh.plotting import figure, output_file, show

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Output to static HTML file
output_file("customization_example.html")

# Create a new plot with custom appearance
p = figure(title="Customization Example",
           x_axis_label='X-axis', y_axis_label='Y-axis',
           width=800, height=400,             # Set plot dimensions
           background_fill_color="lightgrey", # Set background color
           border_fill_color="white")         # Set border color

# Add circle glyph with custom appearance
p.circle(x, y, size=10, color="navy", alpha=0.5, legend_label="Circle")

# Customize axes appearance
p.xaxis.axis_label_text_font_style = "italic"
p.yaxis.axis_label_text_font_style = "italic"
p.xaxis.axis_label_text_font_size = "12pt"
p.yaxis.axis_label_text_font_size = "12pt"

# Customize title appearance
p.title.text_font = "times"
p.title.text_font_size = "20pt"
p.title.align = "center"

# Customize legend appearance
p.legend.title = "Glyph Type"
p.legend.label_text_font = "helvetica"
p.legend.label_text_font_size = "10pt"
p.legend.location = "top_left"

# Show the plot
show(p)
```
